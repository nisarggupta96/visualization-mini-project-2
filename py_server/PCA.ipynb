```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('cars.csv')
```

```python
cols_to_drop = ["name", "torque", "fuel", "seller_type", "transmission", "owner"]
df_1 = df.drop(cols_to_drop, axis=1)
df_1 = df_1.dropna(axis=0)
df_1
```

```python
df_2 = df_1.replace(np.nan,0,regex=True)
df_2 = df_1.replace('',0,regex=True)
df_2
```

```python
df_2['engine'] = df_2['engine'].apply(lambda x: '0' + x.replace(' CC', ''))
df_2['engine'] = df_2['engine'].astype('float')
df_2
```

```python
df_2['max_power'] = df_2['max_power'].apply(lambda x: '0' + x.replace(' bhp', ''))
df_2['max_power'] = df_2['max_power'].astype('float')
df_2
```

```python
df_2['mileage'] = df_2['mileage'].apply(lambda x: '0' + x.replace(' kmpl', '').replace(' km/kg', ''))
df_2['mileage'] = df_2['mileage'].astype('float')
df_2
```

```python
df_2["selling_price"] = pd.qcut(df_2["selling_price"], 4, labels=["Grp 1", "Grp 2", "Grp 3", "Grp 4"])
df_2["selling_price"]
```

```python
scaler = StandardScaler()
scaled_df = df_2.copy()
numeric_columns = ["year", "km_driven", "mileage", "engine", "max_power", "seats"]

scaled_df[numeric_columns] = scaler.fit_transform(scaled_df[numeric_columns])
scaled_df
```

```python
from sklearn.decomposition import PCA

#define PCA model to use
pca_model = PCA(n_components=4)

#fit PCA model to data
res = pca_model.fit_transform(scaled_df[numeric_columns])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()

plt.bar(range(1,len(pca_model.explained_variance_ratio_ )+1), pca_model.explained_variance_ratio_ )
plt.ylabel('Explained variance')
plt.xlabel('Components')
plt.plot(range(1,len(pca_model.explained_variance_ratio_ )+1),
         np.cumsum(pca_model.explained_variance_ratio_),
         c='red',
         label="Cumulative Explained Variance")
plt.legend(loc='upper left')
```

```python
plt.plot(pca_model.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()

#PCA1 is at 0 in xscale
```

```python
print(pd.DataFrame(pca_model.components_[0:2].T, index=scaled_df[numeric_columns].columns))
```

```python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley,s=2)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'green', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
 
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))

myplot(res[:,0:2],np.transpose(pca_model.components_[0:2, :]),list(scaled_df[numeric_columns].columns))
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "random_state": 1,
    "max_iter": 300,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 5):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_df[numeric_columns])
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
